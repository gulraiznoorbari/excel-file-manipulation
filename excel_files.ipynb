```python
# data = pd.read_excel("./PSU-5095.xlsx", sheet_name=None, usecols=["GENE"])
# print(data)

bacteria_file = "PSU-5117.xlsx"
vfdb_data = pd.read_excel(f"./Bacteria_Files/{bacteria_file}", sheet_name="vfdb")
vfdb_data_list = vfdb_data["GENE"].tolist()

ecoli_data = pd.read_excel(f"./Bacteria_Files/{bacteria_file}", sheet_name="ecoli vf")
ecoli_data_list = ecoli_data["GENE"].tolist()

merged_data = list(set(vfdb_data_list + ecoli_data_list))

print(merged_data)
```

```python
def mark_data_in_excel(file_path, merged_data):
    # Load the workbook
    wb = load_workbook(file_path)

    # Get the last column index for each sheet
    for sheet_name in wb.sheetnames:
        # Read the Excel file
        bacteria_data = pd.read_excel(file_path, sheet_name=sheet_name)

        # Get the last column index
        last_column_index = bacteria_data.shape[1] - 1

        # Convert the last column index to column letter (Add 3 for starting from column C)
        last_column_letter = get_column_letter(last_column_index + 3)  

        # Specify the column range for usecols dynamically
        usecols_range = f"C:{last_column_letter}"

        # Read the Excel file with updated column range
        bacteria_data = pd.read_excel(file_path, sheet_name=sheet_name, usecols=usecols_range)

        # Extracting the 6th row of the `bacteria_data` DataFrame and assigning it to the variable `PSU`. This allows us to access and modify the values in that specific row.
        PSU = bacteria_data.iloc[6]

        # Mark the data in the PSU row as "tick" or "cross"
        for i in PSU.index:
            if i in merged_data:
                PSU[i] = '✓'  # Mark as "tick"
            else:
                PSU[i] = '✕'  # Mark as "cross"

        # Apply font color to the tick (✓) and cross (✕) values
        green_font = Font(color="00009900")
        red_font = Font(color="00FF0000")

        # Get the corresponding worksheet
        ws = wb[sheet_name]

        # Write the output to the specified row and column
        output = PSU.values

        row_number = 28
        for column_number, value in enumerate(output, start=3):
            cell = ws.cell(row=row_number, column=column_number)
            cell.value = value
            if cell.value == "✓":
                cell.font = green_font
            elif cell.value == "✕":
                cell.font = red_font
    # Save the changes
    wb.save(file_path)
    wb.close()

    print("Data saved to Excel file.")

file_name = "./VAGs-listing-final.xlsx"
mark_data_in_excel(file_name, merged_data)

```

```python
import openpyxl

def replace_empty_cells_with_cross(file_path):
    # Load the Excel file
    workbook = openpyxl.load_workbook(file_path)

    # Iterate over each sheet in the workbook
    for sheet_name in workbook.sheetnames:
        sheet = workbook[sheet_name]
        max_row = sheet.max_row
        max_col = sheet.max_column

        # Check for empty cells in each row and column within max_row and max_col
        for row in range(1, max_row + 1):
            for col in range(1, max_col + 1):
                if sheet.cell(row=row, column=col).value is None:
                    sheet.cell(row=row, column=col).value = '✕'
                    sheet.cell(row=row, column=col).font = Font(color="00FF0000")

    # Save the modified workbook
    workbook.save(file_path)
    workbook.close()


excel_file_path = './VAGs-listing-final.xlsx'
replace_empty_cells_with_cross(excel_file_path)
```
